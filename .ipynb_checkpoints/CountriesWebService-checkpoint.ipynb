```python
import requests
import csv
```

```python
request = 'https://restcountries.com/v2/all'
```

```python
response = requests.get(request)
```

```python
json_response = response.json()
json_response
```

```python
file_name = 'Countries.csv'
```

```python
with open(file_name, 'w', newline='') as f:
    fh = csv.DictWriter(f, ['name', 'alpha3Code', 'area', 'region', 'population', 'main_currency', 'main_language'])
    fh.writeheader()
    for country in json_response:
        name = country.get('name')
        alpha3Code = country.get('alpha3Code')
        area = country.get('area') if country.get('area') != None  else 'NA'
        region = country.get('region')
        population = country.get('population')
        main_currency = country.get('currencies')[0].get('code') if country.get('currencies') != None  else 'NA'
        main_language = country.get('languages')[0].get('name') if country.get('languages') != None  else 'NA'
        fh.writerow({'name':name, 'alpha3Code':alpha3Code, 'area':area, 'region':region, 'population':population, 'main_currency':main_currency, 'main_language':main_language})
    
```

```python
import pandas as pd
```

```python
countries_data = pd.read_csv(file_name,encoding = 'unicode_escape')
```
